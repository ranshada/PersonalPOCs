```python
meallist=[
  {"meal": "Upma", "calories": 300},
  {"meal": "Poha", "calories": 250},
    {"meal": "Idli", "calories": 200},
    {"meal": "Dosa", "calories": 350},
    {"meal": "Pongal", "calories": 400},
    {"meal": "Biryani", "calories": 600},
    {"meal": "Fried Rice", "calories": 500},
    {"meal": "Khichdi", "calories": 300}
]

def get_max(meallist):
    hightestcalories=0
    print(type(meallist))
    for item in meallist:
        if(item["calories"] > hightestcalories):
            hightestcalories=item["calories"]
    return hightestcalories

print(f'Higest caloried contained meal is   : {get_max(meallist)}')
```

```python
def clean(text):
    text = text.lower()
    text = text.replace("  ", "")
    text=text.strip()
    return text

with open ("sample.txt","r+") as file:
    
    content = file.readlines()
    
    file.seek(0)  # Move the cursor to the beginning of the file
    file.truncate()  # Clear the file before writing cleaned content
    cleaned_content=[clean(line)+"\n" for line in content]
    file.writelines(cleaned_content)
    


```
